# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



Primeiro devemos calcular a integral da função que representa o crescimento do cacto no intervalo de $0$ a $2$: $y = x^5(2 - x)$. Faremos isso da seguinte forma:
$$\int_{0}^{2} x^5(2 - x)dx$$
Expandir a função:
$$x^2(2 - x) = 2x^5 - x^6$$
$$=\int_{0}^{2}(2x^5 - x^6)dx$$
Podemos presseguir dividindo a expressão em duas partes:
$$\int_{0}^{2}2x^5 \ dx \, \text{ &} \int_{0}^{2}x^6 \ dx$$
E então calcular a integral de cada uma individualmente. \
Para $\int_{0}^{2}2x^5 \ dx$:
$$\int_{0}^{2}2x^5 \ dx = 2 \int_{0}^{2} x^5 \ dx = 2\left[\frac{x^6}{6}\right]_{0}^{2} = 2\left(\frac{0^6}{6} - \frac{2^6}{6}\right) = 2\left(\frac{64}{6}\right) = \frac{128}{6}$$
$$ = \frac{64}{3}$$
Agora para a parte restante, $\int_{0}^{2}x^6 \ dx$:
$$\int_{0}^{2}x^6 \ dx = \left[\frac{x^7}{7}\right]_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7}$$
$$ = \frac{128}{7}$$
Agora que temos ambos os valores, basta apenas subtraí-las:
$$\frac{64}{3} - \frac{128}{7} = \frac{64 \cdot 7 - 128 \cdot 3}{21} = \frac{448 - 384}{21}$$
$$= \frac{64}{21}$$
**Portanto, a quantidade que foi acumulada de biomassa no cacto é $= \frac{64}{21}$ m$^3$**

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, lambdify

x = symbols('x')
F = x**5 * (2 - x)

lim_inf = 0
lim_sup = 2

area = integrate(F, (x, lim_inf, lim_sup))

x_vals = np.linspace(0, 2, 1000)
F_lamb = lambdify(x, F)
F_vals = F_lamb(x_vals)

plt.plot(x_vals, F_vals, label='$F(x) = x^5 (2 - x)$', color='r')

plt.fill_between(x_vals, F_vals, color='lightblue', alpha=0.5)

plt.xlim([0, 2])
plt.ylim([0, max(F_vals) + 1])

plt.title(f'Área sob a curva: {area.evalf():.4f}')
plt.legend()

plt.xlabel('x (anos)')
plt.ylabel('F(x) (metros)')

plt.show()

print(f'Área sob a curva: {area.evalf():.4f}')
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

O primeiro passo para resolver o problema é determinar a expressão para a velocidade $v(t)$ com base na aceleração $a(t)$. Depois calculamos a posição $s(t)$, com base na velocidade. \
Primeiro, vamos integrar a aceleração $a(t)$, que se dá por $\cos(πt)$:
$$v(t) = \int a(t) \ dt = \int \cos(\pi t) \ dt$$
Calcular a integral de $\cos(\pi t)$:
$$v(t) = \frac{1}{\pi} \sin(\pi t) + C$$
O próximo passo é encontrar a constante $C$; \
Sabemos que $v(0)$ é igual a $\frac{12}{\pi}$ no instante $t = 0$, então podemos usar isso para obetr o valor, desta forma:
$$v(0) = \frac{1}{\pi}sin(0) + C = \frac{12}{\pi}$$
$$C = \frac{12}{\pi}$$
Assim obtemos a expressão da velocidade:
$$v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{12}{\pi}$$
Agora que concluímos o primeiro passo, obter a expressão da velocidade, podemos partir partir para o segundo passo: encontrar a posição $s(t)$. Para encontrá-la, basta calcular a integral de $v(t)$, deste modo:
$$s(t) = \int v(t) \ dt = \int \left(\frac{1}{\pi} \sin(\pi t) + \frac{12}{\pi} \right) \ dt$$
Seguimos separando as integrais:
$$s(t) = \frac{1}{\pi} \int \sin(\pi t) \ dt + \frac{12}{\pi} \int 1 \ dt$$
Calcular a integral de $\sin(\pi t)$:
$$\int \sin(\pi t) \ dt = -\frac{1}{\pi} \cos(\pi t)$$
Logo, temos:
$$s(t) = \frac{1}{\pi} \left(-\frac{1}{\pi} \cos(\pi t)\right) + \frac{12}{\pi}t
$$
$$s(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{12}{\pi}t + C'$$
Para seguir, vsmos considerar a posição de ínicio, $s(0)$, que é igual a $0$, e aplicar à expressão:
$$s(0) = -\frac{1}{\pi^2} \cos(0) + \frac{12}{\pi} \cdot 0 + C' = 0$$
$$C' = \frac{1}{\pi^2}$$

Por fim, obtemos a expressão para a posição $s(t)$:
$$s(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{12}{\pi}t + \frac{1}{\pi^2}$$

**Por fim, com a expressão da posição obtida, e avaliada para $t = 0$ e $t = 1.5$ (ver gráfico abaixo), temos o nosso resultado: A distância total percorrida nos primeiros $1.5$ segundo é de aproximadamente $5.8309$ unidades de distância.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, cos, sin, pi, integrate, lambdify

t = symbols('t')
a = cos(pi * t)
v = integrate(a, t) + 12/pi
s = integrate(v, t)

a_lamb = lambdify(t, a)
t_vals = np.linspace(0, 1.5, 1000)
a_vals = a_lamb(t_vals)

v_lamb = lambdify(t, v)
v_vals = v_lamb(t_vals)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t_vals, a_vals, label='$a(t) = \cos(\pi t)$', color='b')
plt.title('Aceleração e Velocidade ao longo do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração $a(t)$')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_vals, v_vals, label='$v(t)$', color='r')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade $v(t)$')
plt.legend()

plt.tight_layout()
plt.show()

s_1_5 = s.subs(t, 1.5)
s_0 = s.subs(t, 0)
distance = abs(s_1_5 - s_0)
print(f'Distância total percorrida: {distance.evalf():.4f}')
```

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


**Resultado: a área entre as curvas é igual a:
$$4.0833$$**
Ver código abaixo

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, lambdify

x = symbols('x')
f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3

h = g - f

area = integrate(h, (x, 1, 2))

x_vals = np.linspace(0, 3, 1000)
f_lamb = lambdify(x, f)
g_lamb = lambdify(x, g)
f_vals = f_lamb(x_vals)
g_vals = g_lamb(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='$f(x) = -x^2 + 4x + 1$', color='b')
plt.plot(x_vals, g_vals, label='$g(x) = -x^3 + 7x^2 - 10x + 3$', color='r')

x_fill = np.linspace(1, 2, 1000)
f_fill = f_lamb(x_fill)
g_fill = g_lamb(x_fill)
plt.fill_between(x_fill, f_fill, g_fill, where=(g_fill > f_fill), color='lightgray', alpha=0.5)

plt.xlim([0, 3])
plt.ylim([min(f_vals.min(), g_vals.min()) - 1, max(f_vals.max(), g_vals.max()) + 1])

plt.title(f'Área entre as curvas: {abs(area.evalf()):.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()

print(f'Área entre as curvas no intervalo [1, 2]: {abs(area.evalf()):.4f}')
```

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


Usando o método dos discos, primeiro devemos encontrar os pontos de interseção das duas funções. Fazemos isso igualando as funções, obtendo o ponto em:
$$x = 9$$
Para determinar o volume, seguimos girando a região delimitada pelas funções em torno do eixo $y$. \
A fórmula para o volume no método dos discos é a seguinte:
$$V = \pi \int_{a}^{b} [f(y)]^2 \ dy$$
Senso $a$ e $b$ os pontos de interseção, e sendo $f(y)$  a função que representa a distância entre o eixo de rotação e a função a ser rotacionada. \
Substituir na fórmula:
$$V = \pi \int_{0}^{9} [3 - \sqrt{y}]^2 \ dy$$
O último passo é calcular a integral. Faremos isso da seguinte maneira: \
1. Exaandir o integrando:
$$[3 - \sqrt{y}]^2 = (3 - \sqrt{y})(3 - \sqrt{y}) = 9 - 6\sqrt{y} + y$$
2. Reescrever a integral substituindo a expressão expandida:
$$V = \int_{0}^{9}(9 - 6\sqrt{y} + y) \ dy$$
3. Separar a integral:
$$V = \pi(\int_{0}^{9}9 \ dy - 6 \int_{0}^{9} \sqrt{y} \ dy + \int_{0}^{9} \ y \ dy)$$
4. Calcular cada integral separadamente:
$$\int_{0}^{9}9 \ dy = 9y|_{0}^{9} = 9 \cdot 9 - 9 \cdot 0 = 81$$
$$ \int_{0}^{9} \sqrt{y} \ dy = \int_{0}^{9} y^{1/2} dy = 18$$
$$\int_{0}^{9} \ y \ dy = \left[\frac{y^2}{2}\right]_{0}^{9} = 40.5$$

Por fim, basta apenas substituir e somar os valores:
$$V = \pi(81 - 6 \cdot 18 + 40.5) = \pi(13.5)$$
**Portanto, o volume $V$ do sólido é igual a $13.5 \pi$ unidades cúbicas.**


# Questão 6 - Vale até 3 pontos
Para os problemas a e b, determine a média, usando a integração, para a funções nos intervalos especificados.

a. $f(x) = 8x - 3 + 5e^{2-x}$ no intervalo $[0, 2]$ - Vale 1 ponto

b. Considere que a posição de um objeto em movimento ao longo do eixo $ x $ é descrita pela função de deslocamento $ x(t) = t^3 - 6t^2 + 9t + 1 $, onde $ x(t) $ está em metros e $t$ está em segundos. Determine a velocidade média do objeto no intervalo de tempo de $ t = 1$ segundo até $t = 4$ segundos. - Vale 2 pontos

(b) \
A velocidade média de um objeto é dada pela fórmula:
$$v = \frac{\Delta x}{\Delta t}$$
Para a função em questão, resolveremos da seguinte forma:
1. Calcular $x(4)$:
$$x(4) = 4^3 - 6 \cdot 4^2 + 9 \cdot 4 + 1 = 5$$
2. Calcular $x(1)$:
$$x(1) = 1^3 - 6 \cdot 1^2 + 9 \cdot 1 + 1 = 5$$
3.Calcular $\Delta(x)$:
$$\Delta(x) = x(4) - x(1) = 5 - 5 = 0$$
4. Calcular $\Delta(t)$:
$$\Delta(t) = 4 - 1 = 3$$
5. Calcular a velocidade média substituindo os valores na fórmula:
$$v = \frac{\Delta x}{\Delta t} = \frac{0}{3} = 0$$
**Portanto, a velocidade média no intervalo de tempo é igual a $0$ms.**

```python
import sympy as sp

# (a)
x = sp.symbols('x')
f = 8*x - 3 + 5*sp.exp(2 - x)
media_a = (1 / 2) * sp.integrate(f, (x, 0, 2))
print(f'Média de f(x) no intervalo [0, 2]: {media_a.evalf()}')

# (b)
t = sp.symbols('t')
x_t = t**3 - 6*t**2 + 9*t + 1
x_4 = x_t.subs(t, 4)
x_1 = x_t.subs(t, 1)
delta_x = x_4 - x_1
delta_t = 4 - 1
v_media = delta_x / delta_t
print(f'Velocidade média no intervalo [1, 4]: {v_media}')
```
