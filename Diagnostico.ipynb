```python
import numpy as np
import matplotlib.pyplot as plt

theta = float(input("Digite o ângulo desejado em radianos (0 < θ < π): "))

sin_theta = np.sin(theta)
cos_theta = np.cos(theta)
tan_theta = np.tan(theta)

circle = plt.Circle((0, 0), 1, color='b', fill=False)

fig, ax = plt.subplots()
ax.add_artist(circle)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.plot([0, cos_theta], [0, sin_theta], color='r', label='Seno')
plt.plot([1, cos_theta], [0, 0], color='g', label='Cosseno')
plt.plot([0, tan_theta*cos_theta], [0, tan_theta*sin_theta], color='purple', label='Tangente')

plt.text(0.5, -0.2, f'Ângulo: {theta:.2f} rad', fontsize=10, ha='center')

plt.legend()

plt.title('Círculo Trigonométrico')
plt.xlabel('Cosseno')
plt.ylabel('Seno')
plt.grid(True)
plt.axis('equal')
plt.show()
```
